```python
from datetime import datetime,date
import json
import string
import random
class Food_Ordering_App:
    
    def __init__(self):
        self.Full_Name = ''
        self.Mobile_Number = ''
        self.Email = ''
        self.Address = ''
        self.Password = ''
        self.history=[]
        self.pd={"Full Name":self.Full_Name,"Mobile Number":self.Mobile_Number,"Email":self.Email,"Address":self.Address,"Password":self.Password,"History":self.history}
        
    def Register(self):
        
        self.Full_Name = input("Enter Full Name: ")
        self.Mobile_Number = input("Enter Mobile Number: ")
        self.Email = input("Enter Email: ")
        self.Address = input("Enter Address: ")
        self.Password = input("Enter Password (8 chars) :")
        self.pd={"Full Name":self.Full_Name,"Mobile Number":self.Mobile_Number,"Email":self.Email,"Address":self.Address,"Password":self.Password,"History":[]}
        
        
        if(len(self.Full_Name)*len(self.Mobile_Number)*len(self.Email)*len(self.Password))==0 or len(self.Mobile_Number)!=10 or len(self.Password)<8:
            print("Please enter valid data")
            
        
        else:
            with open('Users.json','r+') as f:

                data = json.load(f)
                
                if self.Mobile_Number in str(data):
                    print("Mobile Number already exists")
                    
                else:
                    data["users"].append(self.pd)
                    f.seek(0)
                    json.dump(data,f)
                    print("Registration Successful")
                    self.User_Menu()
    
    def Login(self):
        self.Full_Name = input("Enter User Name")
        self.Password = input("Enter Password")
        flag=0
        with open('Users.json','r+') as f:
            data = json.load(f)
            for i in data["users"]:
                if self.Full_Name==i["Full Name"] and self.Password==i["Password"]:
                    self.pd=i
                    flag=1
                    print("Logged in successfully")
                
                    
        if flag==0:
            print("Incorrect details")
        else:
            self.User_Menu()
        
        
            
    def Update(self):
        
        flag=0
        with open('Users.json','r+') as f:
            data = json.load(f)
            for i in range(len(data["users"])):
                if self.pd["Mobile Number"]==data["users"][i]["Mobile Number"]:
                    flag=1
                    print('Input new details')
                    
                    self.Full_Name = input("Enter Full Name: ")
                    self.Mobile_Number = input("Enter Mobile Number: ")
                    self.Email = input("Enter Email: ")
                    self.Address = input("Enter Address: ")
                    self.Password = input("Enter Password (8 chars) :")
                    if(len(self.Full_Name)*len(self.Mobile_Number)*len(self.Email)*len(self.Password))==0 or len(self.Mobile_Number)!=10 or len(self.Password)<8:
                        print("Enter valid details")
                    else:
                        data["users"][i]["Full Name"] = self.Full_Name
                        data["users"][i]["Mobile Number"] = self.Mobile_Number
                        data["users"][i]["Email"] = self.Email
                        data["users"][i]["Address"] = self.Address
                        data["users"][i]["Password"] = self.Password
                        print("Details Updated")
                        f.seek(0)
                        json.dump(data,f)
                    
                    
            if flag==0:
                print('Invalid Password')
        self.User_Menu()        
            

    def History(self):
        if self.pd["History"]==[]:
            print("No order history")
        else:
            print('Your order history is: ')
            for i in self.pd["History"]:
                print(i)
        self.User_Menu()
        
        
    def Order(self):
         
        x=eval(input("Enter your choice \n 1. Tandoori Chicken (4 pieces) [INR 240] \n 2. Vegan Burger (1 Piece) [INR 320] \n 3. Truffle Cake (500gm) [INR 900] \n Note:Input should be in the form [1,2] \n"))
        self.menu={1:'Tandoori Chicken (4 pieces) [INR 240]',2:'Vegan Burger (1 Piece) [INR 320]',3:'Truffle Cake (500gm) [INR 900]'}
        self.order_list=[]

        for i in x:
            self.order_list.append(self.menu.get(i))
        print("You have selected the following items")
        for i in self.order_list:
            print(i)
        choice=input("Do you wish to place an order (Y/N) ?")
        if choice=='Y' or choice=="y":
            print("Thankyou! Your order has been placed successfully")
            self.pd["History"].append(self.order_list)
            with open('Users.json','r+') as f:
                      data = json.load(f)
                      for i in range(len(data["users"])):
                            if self.Full_Name==data["users"][i]["Full Name"]:
                                data["users"][i]=self.pd
                      f.seek(0)
                      json.dump(data,f)
        
        else:
            print("Order Cancelled")
        self.User_Menu()
            
            
    def Add_Items(self):
        try:
            
            self.food=input("Enter name of Food Item: ")
            self.quantity=input("Enter Item Quantity: ")
            self.type=input("Enter Food Item Type: ml, gm, piece, etc: ")
            self.price=input("Enter Price of Food Item (in Rs): ")
            self.discount=input("Enter discount (in Rs) on Item: ")
            self.stock=input("Enter stock of Food Item: ")
            self.id=''.join(random.choices(string.ascii_uppercase+string.digits,k=4))
            self.item_details={'Product ID':self.id,'Name':self.food,'Quantity':self.quantity,'Type':self.type,'Price':self.price,'Discount':self.discount,'Stock':self.stock}
            print("Food Item added successfully")
        except Exception as e:
            print("\n!! PLEASE PROVIDE CORRECT INPUTS !!\n ")
            
        with open("Food_Items.json",'r+') as f:
            data=json.load(f)
            data["items"].append(self.item_details)
            f.seek(0)
            json.dump(data,f)
        self.Admin_Menu()
        
        
    def View_Items(self):
        with open("Food_Items.json",'r') as f:
            data=json.load(f)
            if len(data["items"])>0:
                print("The following Items are available: ")
                for i in data["items"]:
                    print(i)
            else:
                print("No items found")
        self.Admin_Menu()
        
    
    def Remove_Items(self):
        self.id=input("Enter the ID of the item to be Deleted: ")
        flag=0
        f=open("Food_Items.json")
        content=json.load(f)
        f.close()
        for i in range(len(content["items"])):
            if self.id==content["items"][i]['Product ID']:
                del content["items"][i]
                flag=1
            
        if flag==1:
            with open("Food_Items.json",'w+') as f:
                f.seek(0)
                json.dump(content,f)
                print("Item removed successfully")
        else:
            print("Item not found")
        self.Admin_Menu()
        
            
    def Edit_Items(self):
        self.id=input("Enter the ID of the item to be Edited: ")
        flag=0
        with open("Food_Items.json",'r+') as f:
            data=json.load(f)
            for i in range(len(data["items"])):
                if self.id==data["items"][i]['Product ID']:
                    self.detail=input("Enter detail to be Edited: ")
                    self.new=input("Enter new detail")
                    data["items"][i][self.detail]=self.new
                    flag=1
                    print("Details edited successfully")
            f.seek(0)
            json.dump(data,f)
            
            
        if flag==0:
            print("Item not found")
        self.Admin_Menu()
        
            
    def Admin_Menu(self):
        v=True
        while v:
            print("\nEnter the Below Options\n")
            print("\t1. ADD FOOD ITEM \n\t2. EDIT FOOD ITEM\n\t3. VIEW FOOD ITEM\n\t4. DELETE FOOD ITEM\n\t5. GO BACK")
            z=input()
            if z=="1":
                self.Add_Items()
            elif z=="2":
                self.Edit_Items()
            elif z=="3":
                self.View_Items()
            elif z=="4":
                self.Remove_Items()
            elif z=="5":
                v=False
            else:
                print("NOT A VALID OPTION ")
               
    def User_Menu(self):
        v=True
        
        while v:
            print("\nEnter the Below Options\n")
            print("\t1. PLACE NEW ORDER\n\t2. ORDER HISTORY\n\t3. UPDATE PROFILE\n\t4. GO BACK")
            z=input()
            if z=="1":
                self.Order()
            elif z=="2":
                self.History()
            elif z=="3":
                self.Update()
            elif z=="4":
                v=False
            else:
                print("NOT A VALID OPTION")
        
    
    def main_menu(self):
        v=True
        print("Welcome to the Food Ordering App")
        while v:
            print("Enter your choice: \n 1.Admin Login \n 2.User Registration \n 3.User Login \n 4.Exit")
            z=input()
            if z=="1":
                self.Admin_Menu()
            elif z=="2":
                self.Register()
            elif z=="3":
                self.Login()
            elif z=="4":
                v=False
            else:
                print("Enter valid data")
            
        
        
```

```python
obj=Food_Ordering_App()
obj.main_menu()
```
